```python
import pandas as pd
df=pd.read_csv('./spam_and_ham_classification.csv')
df=df[:1000]
df
```

```python
import nltk
import re
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus =[]
for i in range (len(df)):
    review=re.sub('[^a-zA-Z]',' ',df['text'][i])
    review=[lemmatizer.lemmatize(word.lower(),pos='v') for word in review.split() if word not in nltk.corpus.stopwords.words("english")]
    corpus.append(" ".join(review))
```

```python
from sklearn.model_selection import train_test_split
y=pd.get_dummies(df['label'])
y=y.iloc[:,0]
X_train, X_test, y_train, y_test = train_test_split(corpus,y,test_size=0.3, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,ngram_range=(1,2))
X_train=cv.fit_transform(X_train).toarray()
X_test=cv.transform(X_test).toarray()
print(X_train)
print(cv.vocabulary_)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detector =MultinomialNB()
spam_detector.fit(X_train,y_train)
```

```python
y_pred=spam_detector.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test,y_pred))
```

```python
import pickle
with open("model.pickle",'wb') as file:
    pickle.dump(spam_detector,file)
```
